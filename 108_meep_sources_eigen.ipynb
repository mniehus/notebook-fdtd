## EigenModeSource

Lineárny alebo plošný zdroj [EigenMode Source](https://meep.readthedocs.io/en/latest/Python_User_Interface/#eigenmodesource) produkuje plošnú vlnu a umožnuje simulovať budenie mikropáskových vedení alebo vlnovodov. Časový priebeh vlny je možné definovať pomocou štandardných typov zdrojov (Custom, Gaussian, Continuous).

```python
%matplotlib inline
from utils.mp_src_plot import *

sources = [mp.EigenModeSource(src=mp.GaussianSource(.1, width=5),
                center=mp.Vector3(0,0,0),
                size=mp.Vector3(y=25), 
                direction=mp.AUTOMATIC)]
                     
sim = mp.Simulation(cell_size=area, boundary_layers=[mp.PML(5.0)],
                geometry=[], sources=sources, resolution=2)

sim.reset_meep() 
sim.run(until=50)
data = sim.get_array(center=mp.Vector3(), size=area, component=mp.Ez)
src_plot(dx, dy, data.transpose(),2)
```
