# Skalárne pole

## Vlastnosti

V matematike a fyzike je skalárne pole funkciou $f(x,y,z)$ priraďujúce skalárnu hodnotu (t.j. určenú jediným čiselným údajom) v každom bode priestoru $(x,y,z)$. Hodnoty skalárneho poľa môžu byť reálne alebo komplexné. Príkladom skalárneho poľa je napríklad v meteorológii teplota, tlak, hustota nebo vlhkosť vzduchu, v elektrotechnike potenciál.

## Numerický výpočet

Pre vytvorenie skalárneho poľa musíme v *numpy* vygenerovať pole 2D alebo 3D hodnôt (definičný obor), nad ktorým skalárne pole zadefinujeme. 

    dx = np.linspace(-1.0, 1.0, 30)  # rozmer a pocet vzoriek v smere osi x
    dy = np.linspace(-1.0, 1.0, 30)  # a v smere osi y
    x,y = np.meshgrid(dx, dy)        # 2D mriezka hodnot

Funkcia \textit{meshgrid()} vygeneruje polia x,y, tieto obsahujú hodnoty, ktoré reprezentujú mriežku hodnôt $(x_0, y_0), (x_0, y_1) \dots$. V nasledujúcom príklade vytvorme a zobrazme skalárne pole v 2D:

\begin{equation*}
f(x,y) = x^2 + \frac{y^3}{2}
\end{equation*}

<!-- MEDSKIP -->

Pre vizualizáciu skalárneho poľa môžeme použiť metódy *contour()* a *plot_surface()* z knižnice *matplotlib*. 3D vizualizácia zobrazuje hodnoty skalárneho poľa nad súradnicami v 2D. 
<!-- MEDSKIP -->

```python
%reset -f 
%matplotlib inline

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

dx = np.linspace(-1.0, 1.0, 30)      # rozsahy suradnic   
dy = np.linspace(-1.0, 1.0, 30) 
x,y = np.meshgrid(dx, dy)            # polia suradnic (xi, yi) ...

def f(x,y):                          # defincia funkcie 2D skalarne pole
    return (x**2 + y**3/2) 

fig = plt.figure(figsize=(8, 7))     # zobrazenie skalarneho pola
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, f(x,y), alpha=0.7, cmap='jet') 
ax.contour(x, y, f(x,y), 10, colors="k", offset=-0.50)
plt.xlabel('X')
plt.ylabel('Y')
plt.show() 
```
