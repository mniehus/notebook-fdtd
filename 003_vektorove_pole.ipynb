# Vektorové pole

## Vlastnosti

Vektorové pole je v matematike a fyzike funkcia priraďujúca každému bodu priestoru vektor (t.j. veličinu, ktorá má veľkosť a smer). Vo fyzike sa vektorové pole používa na popis toho, ako sa daná vektorová veličina mení v priestore od bodu k bodu, napríklad v hydrodynamike rýchlosť kvapaliny, v elektronike elektrické alebo magnetické pole, v metorológii smer a rýchlosť vetra.

V 3D priestore je vektorové pole definované ako

\begin{equation}
\mathbf{F}(x,y,z) =  F_x(x,y,z) \, \mathbf{i}  +  F_y(x,y,z) \, \mathbf{j}  +  F_z(x,y,z) \, \mathbf{k}  
\end{equation}


## Numerický výpočet

Vektorové pole $\mathbf{F}(x,y)$ v 2D môžeme zadefinovať pomocou skalárnych funkcií $F_x(x,y)$ a $F_y(x,y)$ pre jednotlivé zložky poľa

\begin{equation*}
\mathbf{F}(x,y) =  F_x(x,y) \, \mathbf{i} +  F_y(x,y) \, \mathbf{j} =  \left( -y \, (1-x-y)+ \frac{x}{4} \right) \, \mathbf{i}  +  \left(2 -x^2 + \frac{y}{4} \right) \, \mathbf{j} 
\end{equation*}

<!-- MEDSKIP -->
Pre zobrazenie vektorového poľa môžeme použiť metódu *quiver()* z knižnice *matplotlib*
<!-- MEDSKIP -->

```python
%reset -sf 
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

dx = np.linspace(-1.0, 1.0, 10)             # rozsahy suradnic
dy = np.linspace(-1.0, 1.0, 10)
x, y = np.meshgrid(dx, dy)                  # definicny obor vektoroveho pola

[Fx, Fy] = [ -y*(1-x-y)+ x/4, 2-x**2+y/4 ]  # zlozky vektoroveho pola

fig = plt.figure(figsize=(5, 5))
ax = fig.add_subplot(111)
ax.quiver(dx, dy, Fx, Fy, color='r')
plt.xlabel('x')
plt.ylabel('y')
plt.axis('equal')
plt.title('Vektorove pole')
plt.grid()
plt.show()
```
