## GaussianSource

Zdroj [*GausianSource*](https://meep.readthedocs.io/en/latest/Python_User_Interface/#continuoussource) reprezentuje zdroj elektromagnetickej energie v tvare deriv√°ciu Gassoveho impulzu s nulovou strednou hodnotou. 

```python
%matplotlib inline
from utils.mp_src_plot import *

sources = [ mp.Source(mp.GaussianSource(frequency=0.1, width=5),
                component=mp.Ez, 
                center=mp.Vector3(0,0,0) ) ]
                     
sim = mp.Simulation(cell_size=area, 
                boundary_layers=[mp.PML(5.0)],
                geometry=[], 
                sources=sources, 
                resolution=2)

sim.run(until=50)

data = sim.get_array(center=mp.Vector3(), 
                size=area, 
                component=mp.Ez)

src_plot(dx, dy, data.transpose(), 2)
```
